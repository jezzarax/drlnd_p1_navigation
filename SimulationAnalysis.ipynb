```python
import pandas as pd
import json
import seaborn as sb
import numpy as np

sb.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
ddqn_huber = []
with open("ddqn_training_huber.txt", 'r') as df:
    ddqn_huber = json.load(df)
    
dqn_huber = []
with open("dqn_training_huber.txt", 'r') as df:
    dqn_huber = json.load(df)
```

```python
ddqn_huber_df = pd.DataFrame(ddqn_huber).T
dqn_huber_df = pd.DataFrame(dqn_huber).T
```

```python
full_aggs = pd.DataFrame({
    "ddqn_min": ddqn_huber_df.min(axis=1),
    "ddqn_avg": ddqn_huber_df.mean(axis=1),
    "ddqn_max": ddqn_huber_df.max(axis=1),
    "dqn_min": dqn_huber_df.min(axis=1),
    "dqn_avg": dqn_huber_df.mean(axis=1),
    "dqn_max": dqn_huber_df.max(axis=1)
})
```

```python
sb.lineplot(data = full_aggs.rolling(window=100).mean())
```
